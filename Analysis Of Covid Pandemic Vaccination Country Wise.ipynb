### Droping Columns Which Are Not Important

```python
df.drop(columns = 'iso_code', inplace = True)
df.drop(columns = 'daily_vaccinations_raw', inplace = True)
df.drop(columns = 'source_website', inplace = True)
df.drop(columns = 'total_vaccinations_per_hundred', inplace = True)
df.drop(columns = 'people_vaccinated_per_hundred', inplace = True)
df.drop(columns = 'people_fully_vaccinated_per_hundred', inplace = True)
df.drop(columns = 'daily_vaccinations_per_million', inplace = True)
```

```python
df.isnull().sum()
```

### Filling NAN Values With Mean (Avg Values Or The Most Common Values In Dataset Column Wise)

```python
df['total_vaccinations'].fillna(df['total_vaccinations'].mean(),inplace= True)
df['people_fully_vaccinated'].fillna(df['people_fully_vaccinated'].mean(),inplace= True)
df['daily_vaccinations'].fillna(df['daily_vaccinations'].mean(),inplace= True)
df['people_vaccinated'].fillna(df['people_vaccinated'].mean(),inplace= True)
```

```python
df.isnull().sum()
```

### The information regarding the columns.
Country: Representing different countries.

Date: Representig particular date.

Total_vaccinations: Showing total no. of various vaccination present in the particular country.

people_vaccinated: How many people vaccinated in particular country but not completely vaccinated.

people_fully_vaccinated: Showing how many people got fully vaccinated.

Complete_medication: Representig how many people got complete medication kit country wise.

daily_vaccinations: Showing how many people got vaccinations on daily basis.

vaccines: Name of different type of the vaccines available country wise.

source_name: Representing the name of the sources from where the vaccinations are being made available.
